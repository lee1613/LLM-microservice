```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load the .env file
load_dotenv()

# Fallback manual parse if load_dotenv fails in certain environments
api_key = os.environ.get('VULTR_SERVERLESS_INFERENCE_API_KEY')
if not api_key and os.path.exists('../.env'):
    with open('../.env') as f:
        for line in f:
            if 'VULTR_SERVERLESS_INFERENCE_API_KEY' in line:
                api_key = line.split('=')[1].strip().strip('"').strip("' ")

print(f"API Key found: {api_key is not None}")
```

```python
client = OpenAI(
    base_url="https://api.vultrinference.com/v1",
    api_key=api_key
)

response = client.chat.completions.create(
    model="vultr/got-qwen-120b-normalize",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello! Testing Vultr Serverless Inference."}
    ]
)

print(response.choices[0].message.content)
```
